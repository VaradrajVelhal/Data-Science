## Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = sns.load_dataset('tips')
df
```

```python
df['size'].unique()
```

```python
plt.subplot(1,2,1)
sns.histplot(df['total_bill'], kde=True)
plt.subplot(1,2,2)
sns.histplot(df['tip'],kde=True)
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='hex')
```

```python
sns.pairplot(df,hue='size',palette='rainbow')
```

```python
sns.countplot(x = df['sex'],hue=df['smoker']) 
```

```python
sns.barplot(x = df['sex'],y = df['total_bill'])
```

```python
sns.barplot(x = df['sex'],y = df['tip'],estimator=np.sum)
```

```python
sns.boxplot(x='tip',y='day',data=df,palette='rainbow')
```

```python
sns.violinplot(x='day',y='tip',data=df,palette='rainbow')
```

```python
sns.stripplot(x='day',y='total_bill',data=df,palette='rainbow')
```

```python
sns.swarmplot(x='day',y='total_bill',data=df)
```

```python
sns.violinplot(x='tip',y='day',data=df,)
sns.swarmplot(x='tip',y='day',data=df, )
```
