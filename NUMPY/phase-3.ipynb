## Advance operation with Business examples

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])

print("==== Zomato sales analysis ==== ")
print("\n Sales data shape", sales_data.shape)
print("\n Sample data for 1st 3 restaurant: ", sales_data[:3])
```

```python
#total sales per year
print(np.sum(sales_data,axis=0))
yearly_total = np.sum(sales_data[:,1:],axis=0)
print(yearly_total)
# minimun sales per restaurent
yearly_min = np.min(sales_data[:,1:],axis=1)
print(yearly_min)
# maximum sales per restaurent
yearly_max = np.max(sales_data[:,1:],axis=1)
print(yearly_max)
# maximum sales per year
per_yearly_max = np.max(sales_data[1],axis=0)
print(per_yearly_max)
#average sales
avg_sales = np.mean(sales_data[:,1:],axis=1)
print(avg_sales)
```

```python
#cumalative sum
cumsum = np.cumsum(sales_data[:,1:],axis=1)
print(cumsum)
plt.figure(figsize=(8,6))
plt.plot(np.mean(cumsum, axis=0))
plt.title("Average cumalative sales across all restaurents ")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
# vector
vector1 = np.array([1,2,3,4])
vector2 = np.array([4,3,2,1])
print("Addition of vectors: ",vector1 + vector2)
print("Multiplication of vectors: ",vector1 * vector2)
print("Dot product of vectors: ",np.dot(vector1,vector2))
```

```python
restaurent_types = np.array(["Biryani", "Chinese", "Pizza", "Burger", "Tea"])
vectorized_upper = np.vectorize(str.upper)
print("Uppercase restaurant types: ", vectorized_upper(restaurent_types))
```
