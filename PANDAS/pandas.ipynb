## Pandas

```python
import pandas as pd
```

```python
s = pd.Series([10,20,30,40,50,60])
s.name = "Calories"
print(s)
```

```python
s.dtype
s.values
```

```python
s.index
```

```python
s[0:3]
```

```python
#iloc -> location based indexing
s.iloc[3]
```

```python
s.iloc[0:3]
```

```python
index = ["apple", "banana", "orange", "grapes", "mango", "kiwi"]
s.index = index
print(s)
print(s["grapes"])
```

```python
#loc -> label based indexing
print(s.loc[["grapes", "mango"]])
#In the label based indexing your start as well as end index is included in the output.
s.loc["banana":"grapes"]
```

```python
fruit_protin = {
    "apple": 5.3,
    "banana": 1.3,
    "orange": 0.9,
    "grapes": 2.6,
    "mango": 0.8,
    "kiwi": 1.1,
    "avacado": 2.0
}
s2 = pd.Series(fruit_protin,name="Protein")
print(s2)
```

```python
# conditional selection
s2[s2>1]
s2[(s2>0.5) & (s2<=2)] #&
s2[~(s2 > 1)]
```

```python
#Modifying the values in the series
s2["mango"] = 6.09
s2
```

## Data Frame

```python
import numpy as np
#data frame
data = {
    "name": ["Alice", "Bob", "Charlie", "David", "Eve"],
    "age": [25, 30, np.nan, 40, 45],
    "department": ["HR", "IT", "Finance", "Marketing", "Sales"],
    "salary": [50000, 60000, 30000, 80000, 90000,]
}
```

```python
data
```

```python
df = pd.DataFrame(data)
df
```

```python
df.head(2)
```

```python
#loc and iloc also works in data frame
df.iloc[1:3]
```

```python
df.loc[1:3, ["name", "department"]]
```

```python
#individual column selection
df["name"]
```

```python
#drop a column
df.drop("age", axis=1 ,inplace=True) #use inplace if you want to make changes in original data frame
df
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
# Broad casting
df["salary"] = df["salary"] + 5000
df
```

```python
#renaming the column name
df.rename(columns = {'department': 'dept'} , inplace = True)
df
```

```python
df['dept'].unique()
```

```python
df["dept"].value_counts()
```

```python
df["promoted_salary"] = df['salary'] * 10
df
```

```python
#data cleaning
df.isnull().sum()
```
