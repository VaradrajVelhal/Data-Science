```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,5,11)
x
y = x ** 2
y
```

```python
plt.plot(x,y)
plt.title('Hello Matplotlib')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
```

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,4)
plt.plot(x,y)
plt.subplot(2,2,3)
plt.plot(y,x)
plt.subplot(2,2,2)
plt.plot(x**2,x**2)
```

```python
#object oriented method
fig = plt.figure()
axis1 = fig.add_axes([0.1,0.1,1,1])
axis1.plot(x,y)
axis2 = fig.add_axes([0.2,0.4,0.3,0.4])
axis2.plot(y,x)
```

```python
plt.scatter(x,y)
```

```python
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,1,1])
ax.plot(x,y)
plt.savefig('myplot.png')
```
